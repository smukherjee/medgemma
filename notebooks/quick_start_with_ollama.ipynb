This notebook demonstrates how to run MedGemma locally using [Ollama](https://ollama.com/). This provides completely local inference without requiring GPU resources, API tokens, or internet connectivity for inference.

MedGemma is a collection of Gemma 3 variants that are trained for performance on medical text and image comprehension. This Ollama version focuses on text-only medical Q&A capabilities.

Learn more about MedGemma at the [HAI-DEF developer site](https://developers.google.com/health-ai-developer-foundations/medgemma).

Configure the MedGemma model to use throughout this notebook. You can change this to use different variants if available.

```python
# Model configuration - change this to use different MedGemma variants
MODEL_NAME = "alibayram/medgemma:4b"  # @param {type: "string"}

print(f"üìã Using model: {MODEL_NAME}")
print(f"üì¶ Model size: ~2.4GB download")
print(f"üè• Capabilities: Text-only medical Q&A")
```

## Prerequisites and Setup

### Installation Requirements
1. Install Ollama: `brew install ollama` (macOS) or follow [installation instructions](https://ollama.com/download) for other platforms
2. The model will be automatically downloaded during setup (~2.4GB)

### Advantages of Ollama Approach:
- **Privacy-focused**: All inference runs locally
- **No authentication**: No API tokens or accounts needed
- **Offline capability**: Works without internet after initial setup
- **Resource efficient**: Runs on CPU, no GPU required

**Note**: The Ollama model may have different capabilities compared to the official Hugging Face models and currently supports text-only inference.

## Install Dependencies

```python
# Install required Python packages 
# Note: Using minimal dependencies to avoid Python 3.13 compatibility issues
import sys
import subprocess

def install_package(package):
    """Install a package using subprocess to avoid compatibility issues"""
    try:
        result = subprocess.run([sys.executable, "-m", "pip", "install", "--quiet", package], 
                              capture_output=True, text=True)
        if result.returncode == 0:
            print(f"‚úÖ {package} installed successfully")
        else:
            print(f"‚ö†Ô∏è  Issue installing {package}: {result.stderr}")
    except Exception as e:
        print(f"‚ùå Error installing {package}: {e}")

# Install essential packages
packages = ["pillow"]  # Removed requests due to Python 3.13 compatibility issues
for package in packages:
    install_package(package)

print(f"\nüìã Python version: {sys.version}")
print("‚úÖ Dependencies setup completed!")
print("üí° Note: Using urllib instead of requests for better Python 3.13 compatibility")
```

## Setup Ollama and MedGemma Model

```python
import subprocess
import sys
import os

def setup_ollama_medgemma(model_name=MODEL_NAME):
    """Setup Ollama and ensure medgemma model is available"""
    try:
        # Check if Ollama is installed
        result = subprocess.run(["ollama", "list"], capture_output=True, text=True)
        if result.returncode != 0:
            print("‚ùå Ollama is not installed or not in PATH.")
            print("Please install Ollama from: https://ollama.com/download")
            return False
            
        print("‚úÖ Ollama is installed")
        
        # Check if medgemma model is available
        if model_name not in result.stdout:
            print(f"üì• Pulling {model_name} model...")
            print("‚è≥ This may take several minutes (downloading ~2.4GB)")
            pull_result = subprocess.run(["ollama", "pull", model_name], 
                                       capture_output=True, text=True)
            if pull_result.returncode != 0:
                print(f"‚ùå Failed to pull model: {pull_result.stderr}")
                return False
            print("‚úÖ Model pulled successfully")
        else:
            print(f"‚úÖ {model_name} model is available")
            
        return True
        
    except FileNotFoundError:
        print("‚ùå Ollama is not installed. Please install it from: https://ollama.com/download")
        return False

# Setup Ollama and MedGemma
print(f"üöÄ Setting up Ollama with {MODEL_NAME}...")
ollama_ready = setup_ollama_medgemma()

if ollama_ready:
    print("\nüéâ Setup completed successfully!")
    print("You can now run medical Q&A queries with MedGemma.")
else:
    print("\n‚ùå Setup failed. Please check the error messages above.")
```

## Text-Only Medical Q&A with MedGemma

The MedGemma model running through Ollama supports comprehensive medical text analysis and question-answering capabilities.

```python
from IPython.display import Markdown, display
import re
import time

def clean_ollama_output(raw_output):
    """Clean terminal control characters from Ollama output"""
    if not raw_output:
        return ""
    
    # Join all lines
    text = '\n'.join(raw_output) if isinstance(raw_output, list) else str(raw_output)
    
    # Remove ANSI escape sequences and terminal control characters
    ansi_escape = re.compile(r'\x1B(?:[@-Z\\-_]|\[[0-?]*[ -/]*[@-~])')
    text = ansi_escape.sub('', text)
    
    # Remove common terminal control sequences
    text = re.sub(r'\[?\?25[hl]', '', text)  # Cursor visibility
    text = re.sub(r'\[?\?2026[hl]', '', text)  # Terminal modes
    text = re.sub(r'\[\d*[GK]', '', text)  # Cursor positioning
    text = re.sub(r'\[2K\[1G', '', text)  # Clear line and move cursor
    text = re.sub(r'‚†ô|‚†π|‚†∏|‚†º|‚†¥|‚†¶|‚†ß|‚†á|‚†è|‚†ã', '', text)  # Spinner characters
    
    # Clean up extra whitespace
    text = re.sub(r'\n\s*\n', '\n\n', text)  # Multiple newlines
    text = text.strip()
    
    return text

def run_ollama_medgemma(prompt, model=MODEL_NAME):
    """Run inference with MedGemma via Ollama using cleaned shell magic"""
    if not ollama_ready:
        return "‚ùå Ollama is not ready. Please run the setup cell first."
    
    print(f"üîÑ Running inference with {model}...")
    print("‚è≥ This may take 10-30 seconds...")
    
    try:
        # Use shell magic to call Ollama
        result = get_ipython().run_line_magic('sx', f'ollama run {model} "{prompt}"')
        
        if result:
            cleaned_response = clean_ollama_output(result)
            if cleaned_response:
                print("‚úÖ Inference completed!")
                return cleaned_response
            else:
                return "‚ùå Empty response after cleaning"
        else:
            return "‚ùå No response from Ollama"
            
    except Exception as e:
        return f"‚ùå Error: {str(e)}"

# Example medical questions for demonstration
questions = [
    "What is pneumonia?",
    "How do you differentiate bacterial from viral pneumonia?",
    "What are the main symptoms of myocardial infarction?"
]

print("üè• MedGemma Medical Q&A via Ollama")
print("=" * 50)

for i, question in enumerate(questions, 1):
    print(f"\nüìã Question {i}: {question}")
    response = run_ollama_medgemma(question)
    
    display(Markdown(f"""
### Question {i}
**{question}**

**MedGemma Response:**

{response}

---
"""))
    
    # Small pause between questions
    time.sleep(2)

print("\n‚úÖ All queries completed!")
```

## Medical Image Analysis (Experimental)

**Note**: The current MedGemma model through Ollama primarily supports text-only inference. This section demonstrates text-based medical guidance for image analysis scenarios.

```python
import base64
import json
from PIL import Image
from IPython.display import Image as IPImage, display
import urllib.request
import urllib.error

def analyze_image_with_ollama(image_path, prompt="Describe this medical image", model=MODEL_NAME):
    """
    Provide text-based medical guidance for image analysis.
    Note: True multimodal inference may not be supported.
    """
    if not ollama_ready:
        return "‚ùå Ollama is not ready. Please run the setup cell first."
    
    print(f"üñºÔ∏è  Analyzing medical image scenario with {model}...")
    print("‚ö†Ô∏è  Note: Providing text-based medical guidance (multimodal inference may not be supported)")
    
    try:
        # Provide comprehensive text-based medical guidance
        text_prompt = f"""As a medical expert, provide detailed guidance about {prompt.lower().replace('this', 'a')} 
        and what medical professionals typically look for in such examinations. 
        Focus on common findings, diagnostic criteria, and clinical significance. 
        Be specific about what radiologists examine when analyzing medical images."""
        
        # Use shell magic for inference
        result = get_ipython().run_line_magic('sx', f'ollama run {model} "{text_prompt}"')
        
        if result:
            cleaned_response = clean_ollama_output(result)
            if cleaned_response:
                print("‚úÖ Text-based medical guidance completed!")
                return f"**[Text-only medical guidance - Multimodal not supported]**\n\n{cleaned_response}"
            else:
                return "‚ùå Empty response from model"
        else:
            return "‚ùå No response from Ollama"
            
    except Exception as e:
        return f"‚ùå Error processing request: {str(e)}"

# Download example chest X-ray image using urllib (Python 3.13 compatible)
image_path = "Chest_Xray_PA_3-8-2010.png"

if not os.path.exists(image_path):
    image_url = "https://upload.wikimedia.org/wikipedia/commons/c/c8/Chest_Xray_PA_3-8-2010.png"
    print("üì• Downloading example chest X-ray...")
    
    try:
        # Use urllib instead of requests for better Python 3.13 compatibility
        urllib.request.urlretrieve(image_url, image_path)
        print("‚úÖ Image downloaded successfully")
    except Exception as e:
        print(f"‚ùå Failed to download image: {e}")
        print("üí° You can manually download the image from:")
        print(f"   {image_url}")
        image_path = None

# Analyze the image scenario
if image_path and os.path.exists(image_path):
    analysis_prompt = "chest X-ray analysis focusing on identifying any notable findings"
    result = analyze_image_with_ollama(image_path, analysis_prompt)
    
    # Display results
    print("\n" + "=" * 60)
    print("üè• MEDICAL IMAGE ANALYSIS")
    print("=" * 60)
    
    display(Markdown(f"**Analysis Request:** {analysis_prompt}"))
    display(IPImage(filename=image_path, height=300))
    display(Markdown(f"**MedGemma Analysis:**\n\n{result}"))
else:
    print("‚ùå Could not load image for analysis")
    display(Markdown("**Note:** Image analysis requires downloading an example image. Please check your internet connection or download manually."))
```

## Performance and Usage Tips

### Optimizing Performance:
1. **First query**: May take 30-60 seconds as the model loads into memory
2. **Subsequent queries**: Much faster (10-20 seconds) as model stays loaded
3. **Model memory**: The model stays in memory for ~5 minutes of inactivity

### Best Practices:
- Be specific with medical questions for better responses
- Include relevant context (patient age, symptoms, etc.)
- Use medical terminology when appropriate
- Consider breaking complex questions into smaller parts

### Limitations:
- Text-only inference (no true multimodal support)
- Responses may be shorter than cloud-based models
- May have different training data than official HuggingFace models

## Troubleshooting

### Common Issues:

1. **"Ollama is not installed"**:
   - Install via: `brew install ollama` (macOS) or from https://ollama.com/download
   - Ensure ollama is in your PATH

2. **"Model not found"**:
   - Run the setup cell again to download the model
   - Check internet connection for model download

3. **Long response times**:
   - First query is always slower (model loading)
   - Consider reducing question complexity
   - Ensure sufficient system RAM (4GB+ recommended)

4. **"Empty response"**:
   - Try restarting the Ollama service: `ollama serve`
   - Restart the notebook kernel
   - Check system resources
